# Read a .csv from a URL with Pandas


Target website: https://www.football-data.co.uk/data.php

```python
# reading 1 csv file (E0.csv) from the website with pandas library

df_premier21 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2122/E0.csv')
```

```python
# showing dataframe

df_premier21
```

```python
# rename dataframe columns

df_premier21 = df_premier21.rename(columns={'Date':'date',
                                            'HomeTeam':'home_team',
                                            'AwayTeam':'away_team',
                                            'FTHG': 'home_goals',
                                            'FTAG': 'away_goals'})
```

```python
# show new dataframe
df_premier21
```
